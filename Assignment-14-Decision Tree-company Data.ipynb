# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
# Changing the categorical columns into numeric one 
com1 = pd.get_dummies(com)
```

```python
com1.head()
```

```python
#### Since company is interested to know about the segment or attributes causes high sale. 
com1['Category'] = pd.cut(com1['Sales'], 
                        bins=[0,10,np.inf], 
                        labels=['Low/Mid','High'],
                       include_lowest=True)
com1
```

```python
# the applied category here is Y variable and rest all are X variables, hence applying the LableEncoding to Y.
label_encoder=preprocessing.LabelEncoder()
com1['Category']=label_encoder.fit_transform(com1['Category'])
```

```python
X=com1.iloc[:,1:15]
Y=com1.iloc[:,15]
```

```python
### Now spliting our data into training and testing data set
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=40)
```

```python
### Buliding Descion tree
model=DecisionTreeClassifier(criterion='entropy',max_depth=5)
model.fit(x_train,y_train)
```

```python
### Ploting Tree
tree.plot_tree(model);
```

```python
fn=['CompPrice','Income','Advertising','Population','Price','Age','Education',
    'ShelveLoc_Bad','ShelveLoc_Good','ShelveLoc_Medium','Urban_No','Urban_Yes','US_No','US_Yes']
cn=['Low/Mid', 'High']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (6,6), dpi=900)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
plt.title('Decision tree using Entropy Criteria',fontsize=5)
```

```python
#Predicting on test data
preds = model.predict(x_test) # predicting on test data set 
```

```python
pd.Series(preds).value_counts() # getting the count of each category 
```

```python
pd.crosstab(y_test,preds)## geting 2 way table
```

```python
### Accuracy
np.mean(preds==y_test)
```
